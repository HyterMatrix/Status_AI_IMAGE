```python
from imutils import *
```

```python
image = imread('image.jpg')
show(image)
image.shape
```

```python
for i in range(4):
    image = cv2.pyrDown(image)
    print(image.shape)
    show(image)
```

```python
for i in range(4):
    image = cv2.pyrUp(image)
    print(image.shape)
    show(image)
```

```python
# 
image = imread('image.jpg')
down_image1 = cv2.pyrDown(image)
down_image2 = cv2.pyrDown(down_image1)
up_image = cv2.pyrUp(down_image2)
laplaccion = down_image1-up_image
show(laplaccion)
```
